```python
import pandas as pd
```

```python
cards_df = pd.read_csv(
    "../../data/processed/analise_cartoes.csv"
)  # ou substitua por `df = seu_dataframe`
cards_df.info()
```

# Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
from PIL import Image
TEAM_COL = "clube"
POSITION_COL = "posicao"
MINUTE_COL = "tempo_jogo"

times = ["Fortaleza", "Botafogo-RJ", "Corinthians", "Palmeiras", "Flamengo", "Sao Paulo", "Internacional",
         "Atletico-MG", "Cruzeiro", "Gremio", "Vasco", "Bahia", "Fluminense",
         "Athletico-PR", "Vitoria", "Cuiaba", "Atletico-GO", "Criciuma", "Juventude", "Bragantino", ]

times = ["Corinthians"]

for time in times:

    selected_team = time  # substitua pelo nome do time que quer analisar

    df_team = cards_df[cards_df[TEAM_COL] == selected_team]
    # Pivotar os dados para criar a matriz para o heatmap
    heatmap_data = df_team.pivot_table(
        index=POSITION_COL,  # Posição no eixo Y
        columns=MINUTE_COL,  # Minuto no eixo X
        values="Quantidade_Cartoes",  # Intensidade da cor
        aggfunc="sum",  # Soma os valores de cartões
        fill_value=0  # Preenche valores ausentes com 0
    )

    # Criar o heatmap
    plt.figure(figsize=(10, 10))
    sns.heatmap(
        heatmap_data,
        cmap="viridis",  # Escolha da paleta de cores
        annot=False,  # Mostra os valores no heatmap
        cbar_kws={"label": "Quantidade de Cartões", "orientation":"horizontal"},  # Rótulo da barra de cores
        vmin=0,  # Escala mínima
        vmax=61  # Escala máxima
    )

    # Configurar os rótulos e título
    plt.title(f"Heatmap: Posição x Minuto x Quantidade de Cartões ({selected_team})", fontsize=14)
    plt.xlabel("Minuto do Jogo")
    plt.ylabel("Posição")
    plt.tight_layout()

    # Salvar o gráfico em um arquivo temporário
    temp_filename = f"temp_heatmap_{selected_team}.png"
    plt.savefig(temp_filename)
    plt.close()

    # Abrir a imagem com Pillow e aplicar o crop
    # with Image.open(temp_filename) as img:
    #     # Detectar automaticamente os limites do conteúdo e aplicar o crop
    #     cropped_img = img.crop([58, 38, img.width - 88, img.height - 30])
    #     # Salvar a imagem final com o crop
    #     cropped_img.save(f"heatmap_cartoes_{selected_team}.png")
    #     cropped_img.show()  # Mostrar a imagem cortada
```
